```python
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

The following settings will improve the default style and font sizes for our charts.

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

# TIME

Time is a numeric column. The minimum time is 7 and the maximum time is 178 minutes. Thus we can visualize the distribution of time using a histogram with 9 bins (one for each 20 minutes almost) and a box plot. We'll use plotly to make the chart interactive, but you can create similar charts using Seaborn.

```python
df.Time.describe()
```

```python
fig = px.histogram(df, 
                   x='Time', 
                   marginal='box', 
                   nbins=9, 
                   title='Distribution of Time in Persian Language movies')
fig.update_layout(bargap=0.1)
fig.show()
```

As we can see we have a kind of normal distribution here where most of the movies have the time around the average. It's shown that 868 movies' times are between 80 to 100 minutes. To see more about normal distribution you can check the link here: https://en.wikipedia.org/wiki/Normal_distribution

# YEAR

Year is a numeric column. The minimum year is 1968 and the maximum year is 2020 . Thus we can visualize the distribution of year using a histogram with 53 bins (one for each year) and a box plot. We'll use plotly to make the chart interactive, but you can create similar charts using Seaborn.

```python
fig = px.histogram(df, 
                   x='Year', 
                   marginal='box',
                   nbins = 53,
                   color_discrete_sequence=['red'], 
                   title='Distribution of Year')
fig.update_layout(bargap=0.1)
fig.show()
```

Now it's intresting! We can see that there is a kind of linear relation between number of movies per year and years going up. Despite the sudden increment in year 2000, it's shown that each year we had more movies and the year with maximum movies is 2017 and after that there is a decrement.

# Score

Score is a numeric and continuous column. The minimum socre is 0 and the maximum score is 10 . Thus we can visualize the distribution of socre using a histogram with 20 bins (two for each score) and a box plot. We'll use plotly to make the chart interactive, but you can create similar charts using Seaborn.

```python
fig = px.histogram(df, 
                   x='Score', 
                   marginal='box',
                   nbins = 21,
                   color_discrete_sequence=['green'], 
                   title='Distribution of Score')
fig.update_layout(bargap=0.1)
fig.show()
```

Despite the distribution between 7.8 and 8.2, we can say that distribution of scores is slightly uniform and the count of each column is slightly equal.Strongly between 3.8 to 8.7
